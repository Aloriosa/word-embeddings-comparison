```python
import gensim

# Load Google's pre-trained Word2Vec model.
model = gensim.models.KeyedVectors.load_word2vec_format('./GoogleNews-vectors-negative300.bin', binary=True)
```

```python
type(model)
```

```python
tests = []
with open('./our_test.txt') as file:
    for line in file:
        tests.append([w for i, w in enumerate(line.split("'")) if i % 2])
len(tests), tests[0]
```

```python
model.wv.most_similar(['greece', 'beijing'], ['athens'], 5)
```

```python
%%time
answers = []
for test in tests[:100]:
    a, b, c, d = test
    try:
        [res] = model.wv.most_similar(positive=[b, c], negative=[a], topn=1)
    except KeyError:
        res = [None]
    answers.append(res[0])
```

```python
print(len(answers))
answers[:10]
```
